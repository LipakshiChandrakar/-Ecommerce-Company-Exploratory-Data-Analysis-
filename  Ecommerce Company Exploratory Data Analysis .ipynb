### <span style='color:palevioletred'>4.5 Shortlisting Products based on specific filters

 If filtering is done by discount price, product tag and finding the top 10 maximum discount precentage product then it can be done as:

```python
df_pr = df[df['discounted_price'] > 2000]
df_pr = df_pr[df_pr['discounted_price'] < 3000]
df_pr = df_pr[df_pr['product_tag'] == 'handbags']


df_pr.sort_values(by = 'discount_percent', ascending = False).head(10)
```

 ### <span style='color:palevioletred'>4.6 Finding all the brand names

```python
df['brand_name'].unique()
```

 ### <span style='color:palevioletred'>4.7 Finding the top 10 Brands with most number of products

```python
data = []

for brand_name in df['brand_name'].unique():
    
    c = 0
    for i in df['brand_name']:
        if (i == brand_name):
            c += 1

    data.append([brand_name, c])
    
data = pd.DataFrame(data, columns = ['name','products'])

sorted_data=data.sort_values(by = 'products', ascending = False).head(10)
sorted_data
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
sns.barplot(x='name',y='products',data=sorted_data,palette='magma')
```

### <span style='color:palevioletred'>4.8 Finding the top 10 Most Expensive Brands 

```python
x=df.groupby('brand_name').mean()['marked_price'].sort_values(ascending = False).head(10)
```

### <span style='color:palevioletred'> 4.9 Top Rating Brands

```python
df_pr = df[df['rating'] != 0]

df_pr.groupby('brand_name').mean()['rating'].sort_values(ascending = False)
```

### <span style='color:palevioletred'> 4.10 Finding Correlations in table

```python
df.corr()
```

```python
sns.heatmap(df.corr(),linecolor='white',linewidths=1,annot=True,cmap='flare')
```

### <span style='color:palevioletred'>4.10 Finding number of products in price range

### <span style='color:palevioletred'>4.11 Most rating given by customers

```python
sns.histplot(df['rating'],color='purple',bins=50)
```

### <span style='color:palevioletred'>4.12 Is there any relation between price and discount percent

```python
x=df['marked_price']
y=df['discount_percent']
plt.xlabel('marked_price')
plt.ylabel('discount_percent')
plt.scatter(x,y)
plt.show()
```

### <span style='color:palevioletred'>4.12 Top 10 products sell on ecommerce company

```python
datax = []

for product_tag in df['product_tag'].unique():
    
    c = 0
    for i in df['product_tag']:
        if (i == product_tag):
            c += 1

    datax.append([product_tag, c])
    
datax = pd.DataFrame(datax, columns = ['product_tag','products'])

sorted_datax=datax.sort_values(by = 'products', ascending = False).head(10)
sorted_datax
```
